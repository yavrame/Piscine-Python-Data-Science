```python
import pandas as pd
```

```python
col_names = ['datetime', 'user']
views  = pd.read_csv('../data/feed-views.log', sep='\t', names=col_names)
```

```python
views 
```

```python
views['datetime'] = pd.to_datetime(views.datetime)
```

```python
views
```

```python
views['year'] = pd.DatetimeIndex(views['datetime']).year
views['month'] = pd.DatetimeIndex(views['datetime']).month
views['day'] = pd.DatetimeIndex(views['datetime']).day
views['hour'] = pd.DatetimeIndex(views['datetime']).hour
views['minute'] = pd.DatetimeIndex(views['datetime']).minute
views['second'] = pd.DatetimeIndex(views['datetime']).second
views
```

cut: https://dfedorov.spb.ru/pandas/Разделение%20данных%20в%20Pandas%20с%20помощью%20qcut%20и%20cut.html

```python
# Названия категорий
cut_labels = ["night", 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
# Интервалы деления на категории right=False - правая граница не входит
cut_bins = [0, 4, 7, 11, 17, 20, 24]
views['daytime'] = pd.cut(views.hour, bins = cut_bins , labels = cut_labels, right=False)
views
```

Сделаем столбец user индексом

```python
views = views.set_index('user')
views
```

Посчитаем количество элементов (строк)

```python
views.count()
```

Посчитаем количество элементов в интервалах дня

```python
views['daytime'].value_counts()
```

Отсортируем по заданию

```python
df_sort = views.sort_values(['hour', 'minute', 'second'])
df_sort.head()
```

Найдем максимальное значение в колонке "час" в категории "ночь"

```python
views[views['daytime'] == 'night']['hour'].max()
```

Найдем минимальное значение в колонке "час" в категории "утро"

```python
views[views['daytime'] == 'morning']['hour'].min()
```

Покажем 3 самых ранних часа утра и соответствующие имена пользователей и

Используем метод describe(), чтобы получить базовую статистику для столбцов

Вычислим межквартильный диапазон для часа, извлекая значения из результата метода describe() и сохраняя его в переменной iqr

```python
iqr = views.describe()['hour'].loc['75%'] - views.describe()['hour'].loc['25%']
iqr
```
